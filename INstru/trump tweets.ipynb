```python
#stopWords array
stopWords = ['to', 'the', 'a', 'to', 'is', 'and', 'in', 'of', 'that', 'it', 'be','at', 'this', 'are', 'be', 'for', 'will', 'with', 'at', 'have','on','&amp', 'by']
```

```python
#counter checking the stop words array
print(stopWords)
print(len(stopWords))
```

```python
#relative reference
file = open('words.csv')
```

```python
f = file.read()
```

```python
#total words in the file
array = []
for word in f.split():
    array.append(word)
len(array)
```

```python
#unique words from trumps tweets
unique_words = []
for word in array:
    if word in unique_words:
        continue
    else:
        unique_words.append(word)
len(unique_words)
```

```python
#clean array of words not present in stopWords array
cleanArray = []
for word in array:
    if word in stopWords:
        continue
    else:
        cleanArray.append(word)
```

```python
#checking the length of the cleanArray
len(cleanArray)
```

```python
#dictionary for order of occurence of words
new_dict = dict()
```

```python
for word in cleanArray:
    if word in new_dict:
        new_dict[word] +=1
    else:
        new_dict[word] = 1
```

```python
#printing the dictionary of words
new_dict
```

```python
#order the dictionary according the order of occurence
#use a list for the sorting
sorted_words = sorted(new_dict.items(), key = lambda x: x[1], reverse=True)
```

```python
#convert the list to a dictionary
converted_dict = dict(sorted_words)
converted_dict
```

```python
#slicing the 10 most used words
import itertools
final_dict = dict(itertools.islice(converted_dict.items(), 10))
```

```python
#printing the 10 most used used words dictionary
final_dict
```

```python
#plotting the data using matplotlib
import matplotlib.pyplot as plt
words = list(final_dict.keys())
values = list(final_dict.values())
plt.bar(range(len(final_dict)), values, tick_label=words)
ax = plt.gca()
#rotation of x-axis ticks
ax.tick_params(axis='x', labelrotation = 45)
#title
plt.title("Histogram of Trump's most used words on Twitter against their occurence")
#axis label
plt.xlabel("Words in Trump's Tweets")
plt.ylabel("Number of Occurence")
#saving to a .jpg file
plt.savefig('tweets.jpg')
plt.show()
```
